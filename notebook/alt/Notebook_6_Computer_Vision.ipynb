# Notebook 6 – Computer Vision

Pixelrepräsentation, Filter, Faltung und optional Segmentierung.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
```

## Bild laden und darstellen

```python
from skimage import data
img = data.astronaut()
plt.figure()
plt.imshow(img)
plt.axis("off")
plt.show()

img.shape
```

## Kanaltrennung

```python
r, g, b = img[:,:,0], img[:,:,1], img[:,:,2]
r.shape
```

## Faltung per Hand auf einem 3x3 Patch

```python
patch = r[100:103, 100:103].astype(float)
k = np.array([[0, -1, 0],
              [-1, 5, -1],
              [0, -1, 0]], dtype=float)

conv_value = np.sum(patch * k)
patch, k, conv_value
```

## Faltung auf dem ganzen Bild (Library Vergleich)

```python
from scipy.signal import convolve2d
r_f = convolve2d(r.astype(float), k, mode="same", boundary="symm")

plt.figure()
plt.imshow(np.clip(r_f, 0, 255), cmap="gray")
plt.axis("off")
plt.show()
```
